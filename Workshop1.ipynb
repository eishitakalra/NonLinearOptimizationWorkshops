```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
xlist = np.linspace(-1, 1, 100)
ylist = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(xlist, ylist)

Z = X**2 + Y**2
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
# cp = ax.plot_surface(X, Y, Z)
# cp = ax.plot_surface(X, Y, Z, cmap=plt.get_cmap('hsv'))
cp = ax.plot_wireframe(X, Y, Z)
plt.show()

```

```python
xlist = np.linspace(-1.25, 1.25, 100)
ylist = np.linspace(-1.25, 1.25, 100)
X, Y = np.meshgrid(xlist, ylist)


Z2 = X*X*(4-2.1*X**2+X**4/3) + X*Y + Y*Y*(-4+4*Y*Y)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
# cp = ax.plot_surface(X, Y, Z2)
cp = ax.plot_surface(X, Y, Z2, cmap=plt.get_cmap('hsv'))
# cp = ax.plot_wireframe(X, Y, Z2)
plt.show()
```

```python
fig2, ax2 = plt.subplots()
cp = ax2.contour(xlist, ylist, Z2, 50)

ax2.clabel(cp, inline=True, fmt='%1.0f', fontsize=7)
```
